```python
import networkx as nx
from sympy import binomial
from linkages import *

def laman_comparison(G,d):
    n = G.number_of_nodes()
    e = G.number_of_edges()
    res = d*n-binomial(d+1,2)
    if e < res :return -1
    if e == res:return 0
    if e > res :return 1
    
graph = parallel_4_bar()
nx.draw(graph)
if laman_comparison(graph,DIM) == -1: print("graph is flexible")
```
