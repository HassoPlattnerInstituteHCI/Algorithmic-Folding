```python
import networkx as nx
import itertools
import sympy as sy
    
def laman_comparison(G,d):
    n = G.number_of_nodes()
    e = G.number_of_edges()
    res = d*n-sy.binomial(d+1,2)
    if e < res :return -1
    if e == res:return 0
    if e > res :return 1
    
graph = nx.Graph()
dimension = 2
graph.add_edges_from([(1,2),(2,3),(3,4),(4,1)])
nx.draw(graph)
if laman_comparison(graph,dimension) == -1: print("graph is flexible")
```
